```python
#Drop Duplicate attributes
L =[]
for i in xrange(0,df.columns.values.size-1) : 
    for y in xrange(i+1,df.columns.values.size-1) :
        if(np.array_equal( np.array(df[df.columns[i]]), np.array(df[df.columns[y]]))):
            L.append(df.columns[y])
            
df = df.drop(L, axis=1)
```

```python
learners = [LogisticRegression(), DecisionTreeClassifier(), KNeighborsClassifier(),
            RandomForestClassifier(), GradientBoostingClassifier(), AdaBoostClassifier(), MLPClassifier()]
labels = ['logisticRegression', 'DecisionTree', 'KNN', 'RandomForest', 'GradientBoosting', 
          'AdaBoost', 'NeuralNetwork']
```

```python
#Function to return random parameters of a given classifier
def getRandParams(classifier_name):
    p=dict()
    for key in params[classifier_name].keys():
        liste = params[classifier_name][key]
        p[key]=liste[rand.randint(0,len(liste)-1)]
    return p
#
def getCombinaisonsNumber(classifier_name):
    cpt = 0
    for key in params[classifier_name].keys():
        liste = params[classifier_name][key]
        cpt += len(liste)
    return cpt 
#########################
def getdictSize():
    cpt = 0
    for clf in labels:
        for key in params[clf].keys():
            liste = params[clf][key]
            cpt += len(liste)
    return cpt 
```

```python
def predicts_tocsv(predictions, learner, params):
    cols = [learner+'_'+str(i)+'_'+str(params[i-1]) for i in range(1, len(predictions[0,:])+1)]
    predicts = pd.DataFrame(predictions[:,0:], columns=cols)
    predicts.to_csv(learner+'.csv', sep=',', encoding='utf-8')
    print '____________saved____________'
```

```python
%%time
#for learners with Parameters
scaler = StandardScaler()
cpt = 0
learner_score = pd.DataFrame(columns=['Learner', 'scoreAUC'])
for learner, label in zip(learners, labels):
    dictList = []
    k = 0
    paramSize = getCombinaisonsNumber(label)
    predictions_learner = np.ndarray(shape=(y_test.shape[0], paramSize))
    print label+' >> is processing...'
# Tune Parameters
    for i in range(0, paramSize):
        clfParams = getRandParams(label)
        dictList.append(clfParams.items())
        print clfParams
        learner.set_params(**clfParams)
        learner.fit(X_train, y_train)
        y_pred = learner.predict_proba(X_test)[:, 1]
        score = roc_auc_score(y_test, y_pred)
        print 'Model: ',k,' > has score AUC == ',score
        if(score>=0.6):
            predictions_learner[:,k] = y_pred
            k+=1
            learner_score.loc[i] = [label+'_'+str(k), score]
            cpt+=1
    predicts_tocsv(predictions_learner, label, dictList)      
    print predictions_learner.shape
    print label+' ended processing...'
print cpt, ' Model was generated'
```
